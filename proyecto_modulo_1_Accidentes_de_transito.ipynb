## Eliminamos columnas que no aportan informacion relevante: Cobertura, año, estatus.

```python
print(df['COBERTURA'].value_counts())
print(df['ANIO'].value_counts())
print(df['ESTATUS'].value_counts())
```

```python
#hacemos copia del df original
atus_df = df.copy()
#Eliminamos columnas que no aportan informacion relevante: Cobertura, año, estatus.
atus_df= atus_df.drop(['COBERTURA', 'ANIO','ESTATUS'], axis=1)
```

## Eliminamos renglones que no aportan informacion relevante: 'Certificado cero'.

> Los datos e informes que los Informantes del Sistema proporcionen para fines estadísticos y que provengan de registros administrativos, serán manejados observando los principios de confidencialidad y reserva, por lo que no podrán divulgarse en ningún caso en forma nominativa o individualizada, ni harán prueba ante autoridad judicial o administrativa, incluyendo la fiscal, en juicio o fuera de él.




```python
certif_cero_count = (atus_df == 'Certificado cero').sum()
certif_cero_count
```

```python
certif_cero = atus_df[ atus_df['SEXO'] == 'Certificado cero' ].index
atus_df.drop(certif_cero , inplace=True)
atus_df.reset_index(drop=True, inplace=True)
atus_df.shape
```

##Cambiamos los nombres de la entidad federativa

```python
# Cargamos el catalogo de entidades
entidades ="/content/drive/MyDrive/tc_entidad.csv"
entidades_df= pd.read_csv(entidades, index_col=False)
```

```python
atus_df = entidades_df.merge(atus_df, how="inner")
atus_df= atus_df.drop(["ID_ENTIDAD"], axis=1)
atus_df.rename(columns = {'NOM_ENTIDAD':'ID_ENTIDAD'}, inplace = True)
atus_df.head()
```

## ESTADOS CON MAYOR NÚMERO DE ACCIDENTES CON PEATONES (ATROPELLAMIENTOS)




```python
peatones_df = atus_df.loc[(atus_df['TIPACCID'] == "Colisión con peatón (atropellamiento)") & (atus_df['PEATMUERTO'] == 1) | (atus_df['PEATHERIDO'] == 1 )]
conteos_peatones = peatones_df.ID_ENTIDAD.value_counts()

plt.figure(figsize=(16,7))
sns.barplot(x = conteos_peatones.index[:10], y = conteos_peatones.values[:10])
plt.xticks(rotation= 45, ha="right")
plt.xlabel("Estados")
plt.ylabel("Número de peatones afectados")
plt.title("Los 10 estados más peligrosos para peatones 2021")

plt.show()
```

PORCENTAJE DE PEATONES HERIDOS VS. MUERTOS



```python
peat_herido = ((peatones_df["PEATHERIDO"].sum())/10195)*100
peat_muerto = ((peatones_df["PEATMUERTO"].sum())/10195)*100
total_atropellamiento = len(peatones_df)

labels = 'Heridos', 'Muertos'
sizes = [peat_herido, peat_muerto]


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=40,
       colors = ["lightcoral","brown"])
plt.title("Peatones")
plt.figure(figsize = (50, 50))
plt.show()
```
